## Objective

The AI Recommendation Engine analyzes the predicted infrastructure status of each village and automatically generates development recommendations.

The recommendations include:

- Roads
- Street Lights
- Water Facilities
- Hospitals
- Schools
- Bus Stops
- Estimated Budget
- Expected Infrastructure Improvement

```python
# ==========================================
# Import Libraries
# ==========================================

import pandas as pd

print("Libraries Loaded Successfully ✅")
```

```python
# ==========================================
# Load Village Dataset
# ==========================================

master_df = pd.read_csv("../datasets/processed/village_master_dataset.csv")

master_df
```

# AI Recommendation Rules

The recommendation engine converts infrastructure priority into actionable development suggestions.

These recommendations simulate how a decision support system can assist government authorities in planning rural development.

```python
# ==========================================
# Recommendation Engine
# ==========================================

def recommend(priority):

    if priority == "Critical":

        return (
            "Construct Roads, Install Street Lights, "
            "Build Water Facility, Add Bus Stop, "
            "Improve Drainage"
        )

    elif priority == "Needs Improvement":

        return (
            "Improve Roads, Add LED Street Lights, "
            "Upgrade Water Supply"
        )

    else:

        return (
            "Regular Maintenance and Monitoring"
        )
```

```python
# ==========================================
# Generate Recommendations
# ==========================================

master_df["Recommendation"] = master_df["Priority"].apply(recommend)

master_df
```

# Budget Estimation

Estimated development cost is calculated based on infrastructure priority.

The budget values are approximate and intended for decision-support simulation.

```python
# ==========================================
# Budget Estimation
# ==========================================

def estimate_budget(priority):

    if priority == "Critical":

        return 5000000

    elif priority == "Needs Improvement":

        return 2500000

    else:

        return 500000

master_df["Estimated_Budget"] = master_df["Priority"].apply(
    estimate_budget
)

master_df
```

```python
master_df["Budget (Lakhs)"] = master_df["Estimated_Budget"]/100000

master_df
```

```python
final_report = master_df[[
    "Village",
    "Priority",
    "Recommendation",
    "Budget (Lakhs)"
]]

final_report
```

```python
import os

os.makedirs("../reports", exist_ok=True)

final_report.to_csv(
    "../reports/village_development_report.csv",
    index=False
)

print("✅ Village Development Report Saved Successfully")
```
