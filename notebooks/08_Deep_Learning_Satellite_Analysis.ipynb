# Image Data Preparation

The CNN model requires satellite images for training.

Images are automatically loaded from the training and testing folders.

Each image is resized to 224 × 224 pixels before training.

```python
# ==========================================
# Image Parameters
# ==========================================

IMG_HEIGHT = 224
IMG_WIDTH = 224

BATCH_SIZE = 16

print("Image Size :", IMG_HEIGHT, "x", IMG_WIDTH)
```

```python
# ==========================================
# Dataset Paths
# ==========================================

train_path = "dataset/train"

test_path = "dataset/test"

print("Training Path :", train_path)
print("Testing Path :", test_path)
```

```python
import os

print("Train folders:", os.listdir(train_path))
print("Test folders:", os.listdir(test_path))
```

```python
import os

for folder in ["Critical", "Developed", "Moderate"]:

    train_count = len(os.listdir(f"dataset/train/{folder}"))
    test_count = len(os.listdir(f"dataset/test/{folder}"))

    print(f"{folder}")
    print("Train Images :", train_count)
    print("Test Images  :", test_count)
    print("-"*30)
```

```python
# ==========================================
# Load Image Dataset
# ==========================================

train_dataset = tf.keras.preprocessing.image_dataset_from_directory(

    train_path,

    image_size=(IMG_HEIGHT, IMG_WIDTH),

    batch_size=BATCH_SIZE

)

test_dataset = tf.keras.preprocessing.image_dataset_from_directory(

    test_path,

    image_size=(IMG_HEIGHT, IMG_WIDTH),

    batch_size=BATCH_SIZE

)
```
