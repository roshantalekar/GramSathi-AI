## Objective

The objective of this notebook is to create the master dataset for all study villages.

Each village will become one record in the dataset.

This dataset will later be enriched with infrastructure features and used for Machine Learning and Deep Learning models.

```python
# ==========================================
# Import Libraries
# ==========================================

import pandas as pd
import geopandas as gpd
import numpy as np

print("Libraries Loaded Successfully ✅")
```

```python
# ==========================================
# Study Villages
# ==========================================

villages = pd.DataFrame({

    "Village":[

        "Sadashivgad",

        "Majali",

        "Asnoti",

        "Mudgeri",

        "Kanasgiri",

        "Hotegali",

        "Hankon",

        "Joog"

    ]

})

villages
```

```python
# ==========================================
# Village Master Dataset
# ==========================================

master_df = villages.copy()

master_df["Road_Length_km"] = 0.0

master_df["School_Count"] = 0

master_df["Hospital_Count"] = 0

master_df["Bus_Stop_Count"] = 0

master_df["Building_Count"] = 0

master_df["Water_Count"] = 0

master_df["Road_Density"] = 0.0

master_df["Accessibility_Score"] = 0.0

master_df["Infrastructure_Readiness_Index"] = 0.0

master_df["Priority"] = "Unknown"

master_df
```

# Why Create a Master Dataset?

Instead of training Machine Learning models on raw OpenStreetMap data, the infrastructure information is aggregated at the village level.

Each row in the master dataset represents one village.

Each column represents an infrastructure indicator.

This approach makes the dataset suitable for predictive modeling and decision support.

```python
# ==========================================
# Save Master Dataset
# ==========================================

master_df.to_csv(
    "../datasets/processed/village_master_dataset.csv",
    index=False
)

print("Village Master Dataset Saved Successfully ✅")
```
